#Humanitarian Aid to Turkey from the United States

For years of the Syrian Refugee Crisis/Syrian Civil War

Data soure: [ForeignAssistance.gov](https://www.foreignassistance.gov/explore/country/Turkey)

```
import pandas as pd
sample_turkey_aid = pd.read_excel("turkey_foreign_assistance.xlsx")
print(sample_turkey_aid.sample(10))
```

```
print(list(sample_turkey_aid))
```

```
turkey_aid = sample_turkey_aid[["Accountable_Organization", "Implementing_Organization", "Award_Status", "Award_Total_Estimated_Value", "Award_Start_Date", "Award_End_Date", "Award_Transaction_Value", "Award_Transaction_DAC_Purpose_Code", "Award_Transaction_DAC_Purpose_Code_Name",]]
pd.DataFrame(turkey_aid.sample(10))
```

## Cleaning Data

### Getting proper date range (2010 - present)

```
import datetime
turkey_conflict_aid = turkey_aid[(turkey_aid["Award_Start_Date"]>datetime.date(2010,1,1))]
pd.DataFrame(turkey_conflict_aid.head(10))
```

### Transaction Values:
Making sure all transaction amounts are positive:

```
turkey_conflict_aid["Award_Transaction_Value"]=turkey_conflict_aid["Award_Transaction_Value"].abs()
pd.DataFrame(turkey_conflict_aid.head(10))
```

## Network building:
Linking implementing organizations to aid type:

### Grouping Data:

```
turkey_aid_orgs = pd.DataFrame({'count' : turkey_conflict_aid.groupby(["Implementing_Organization","Award_Transaction_DAC_Purpose_Code_Name"]).size()}).reset_index()
pd.DataFrame(turkey_aid_orgs.head(10))
```

Types of aid to Turkey:

```
turkey_aid.Award_Transaction_DAC_Purpose_Code_Name.drop_duplicates()
```

### Building the actual network:

Organization - Type of aid - Amount

I want to build a network that shows the connection between implementing organizations and their aid type, and also the amount donated per organization for each kind of aid.

### Grouping and adding values summed:

```
turkey_aid_values = pd.DataFrame({'Total Amount' : turkey_conflict_aid.groupby(["Implementing_Organization","Award_Transaction_DAC_Purpose_Code_Name"])["Award_Transaction_Value"].sum()}).reset_index()
pd.DataFrame(turkey_aid_values.head(10))
```

```
import networkx as nx
Gvalues = nx.Graph()
Gvalues.add_nodes_from(turkey_aid_orgs["Implementing_Organization"], node_type="organization")
Gvalues.add_nodes_from(turkey_aid_orgs["Award_Transaction_DAC_Purpose_Code_Name"], node_type="aid")
Gvalues.add_weighted_edges_from(turkey_aid_orgs.values)

nx.write_graphml(Gvalues, "turkey_aid_values.graphml")
```

## Data Visualization:

### Grouping aid by year:
I need to group the types of aid per year, so I will add a column with only a year value for the date.

```
import matplotlib.pyplot as plt
```

```
turkey_conflict_aid["Year"]=turkey_conflict_aid["Award_Start_Date"].map(lambda x: x.year)
pd.DataFrame(turkey_conflict_aid.sample(10))
```

```
turkey_aid_years = pd.DataFrame({'Total_Amount' : turkey_conflict_aid.groupby(["Year"])["Award_Transaction_Value"].sum()}).reset_index()
pd.options.display.float_format = '${:,.2f}'.format
pd.DataFrame(turkey_aid_years)
```

```
turkey_aid_years.to_csv("turkey_aid.csv")
```

```
ax = plt.gca()
turkey_aid_years.plot(kind='bar',x='Year',y='Total_Amount',ax=ax)


plt.show()

plt.savefig("aid_per_year.pdf")
```
